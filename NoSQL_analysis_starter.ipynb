### 1. Which establishments have a hygiene score equal to 20?

```python
pprint(db.establishments.find_one())
```

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {'scores.Hygiene': {'$eq': 20}}
hygiene_fields = {'meta': 0}
#hygiene_fields = {'BusinessName': 1, "BusinessType":1, 'AddressLine2': 1, 'AddressLine3':1, 'LocalAuthorityName':1, 'scores.ConfidenceInManagement': 1,
                #'scores.Structural': 1, 'FHRSID': 1}
    
# Use count_documents to display the number of documents in the result
hygiene_count = establishments.count_documents(hygiene_query)
hygiene_count

# Display the first document in the results using pprint
hygiene_results = establishments.find(hygiene_query, hygiene_fields)

for i in range(1):
    pprint(hygiene_results[i])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.json_normalize(hygiene_results)

# Display the number of rows in the DataFrame
len(hygiene_df)

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
rate_gte4_query = {'LocalAuthorityName': {'$regex': "London"}, 'RatingValue': {'$gte': '4'}}
rate_gte4_fields = {'meta': 0}

# Use count_documents to display the number of documents in the result
rate_gte4_count = establishments.count_documents(rate_gte4_query)
rate_gte4_count

# Display the first document in the results using pprint
rate_gte4_results = establishments.find(rate_gte4_query, rate_gte4_fields)

for i in range(1):
    pprint(rate_gte4_results[i])
```

```python
# Convert the result to a Pandas DataFrame
rate_gte4_df = pd.json_normalize(rate_gte4_results)
# Display the number of rows in the DataFrame
len(rate_gte4_df)
# Display the first 10 rows of the DataFrame
rate_gte4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query ={'RatingValue': {'$eq': '5'}, 'geocode.latitude': {'$gte': (latitude - degree_search), '$lte': (latitude + degree_search)},
       'geocode.longitude': {'$gte': (longitude - degree_search), '$lte': (longitude + degree_search)}}

fields = {'meta': 0}
sort = [('scores.Hygiene', -1)]
limit = 5

results = list(establishments.find(query, fields).sort(sort).limit(limit))

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
top5_df = pd.json_normalize(results)
top5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}

sort_values = {'$sort': {'count': -1 }}

pipeline = [match_query, group_query, sort_values]

final_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in result: ", len(final_results))

# Print the first 10 results
pprint(final_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(final_results)

# Display the number of rows in the DataFrame
print('Rows in DataFrame: ', len(result_df))

# Display the first 10 rows of the DataFrame
result_df.head()
```
